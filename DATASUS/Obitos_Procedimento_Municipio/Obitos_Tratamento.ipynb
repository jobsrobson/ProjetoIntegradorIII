# 2019

```python
# Importando os dados Brutos de 2019

df_2019_1 = pd.read_csv('C:\\Users\\jobsr\\OneDrive\\Documentos\\GitHub\\ProjetoIntegradorIII\\DATASUS\\Obitos_Procedimento_Municipio\\Brutos\\2019_1_O.csv', sep=';', encoding='latin-1')
df_2019_2 = pd.read_csv('C:\\Users\\jobsr\\OneDrive\\Documentos\\GitHub\\ProjetoIntegradorIII\\DATASUS\\Obitos_Procedimento_Municipio\\Brutos\\2019_2_O.csv', sep=';', encoding='latin-1')
df_2019_3 = pd.read_csv('C:\\Users\\jobsr\\OneDrive\\Documentos\\GitHub\\ProjetoIntegradorIII\\DATASUS\\Obitos_Procedimento_Municipio\\Brutos\\2019_3_O.csv', sep=';', encoding='latin-1')
df_2019_4 = pd.read_csv('C:\\Users\\jobsr\\OneDrive\\Documentos\\GitHub\\ProjetoIntegradorIII\\DATASUS\\Obitos_Procedimento_Municipio\\Brutos\\2019_4_O.csv', sep=';', encoding='latin-1')
```

```python
# Drop das colunas "Acompanhamento_Intercorrencias_Pre_Pos_Transplante" e "Total"
df_2019_1.drop(columns=['Acompanhamento_Intercorrencias_Pre_Pos_Transplante', 'Total'], inplace=True)
df_2019_2.drop(columns=['Acompanhamento_Intercorrencias_Pre_Pos_Transplante', 'Total'], inplace=True)
df_2019_3.drop(columns=['Acompanhamento_Intercorrencias_Pre_Pos_Transplante', 'Total'], inplace=True)
df_2019_4.drop(columns=['Acompanhamento_Intercorrencias_Pre_Pos_Transplante', 'Total'], inplace=True)


# Gerar uma nova coluna com o ano e o trimestre para cada tabela de 2019
df_2019_1['Ano'] = 2019
df_2019_1['Trimestre'] = 1

df_2019_2['Ano'] = 2019
df_2019_2['Trimestre'] = 2

df_2019_3['Ano'] = 2019
df_2019_3['Trimestre'] = 3

df_2019_4['Ano'] = 2019
df_2019_4['Trimestre'] = 4


# Gerar um código primário unindo as colunas Ano, Trimestre, Código do Município. Essa coluna será usada como chave primária para mesclar as tabelas. Criar uma nova coluna com o nome "UID"
df_2019_1['UID'] = df_2019_1['Ano'].astype(str) + df_2019_1['Trimestre'].astype(str) + df_2019_1['Codigo_Municipio'].astype(str)
df_2019_2['UID'] = df_2019_2['Ano'].astype(str) + df_2019_2['Trimestre'].astype(str) + df_2019_2['Codigo_Municipio'].astype(str)
df_2019_3['UID'] = df_2019_3['Ano'].astype(str) + df_2019_3['Trimestre'].astype(str) + df_2019_3['Codigo_Municipio'].astype(str)
df_2019_4['UID'] = df_2019_4['Ano'].astype(str) + df_2019_4['Trimestre'].astype(str) + df_2019_4['Codigo_Municipio'].astype(str)



# Mesclar as tabelas em uma única tabela
df_2019 = pd.concat([df_2019_1, df_2019_2, df_2019_3, df_2019_4], ignore_index=True)

# Mover as três últimas colunas para o início da tabela concatenada do ano
cols = df_2019.columns.tolist()
cols = cols[-3:] + cols[:-3]
df_2019 = df_2019[cols]

# Salvar a tabela concatenada em um arquivo csv
df_2019.to_csv('C:\\Users\\jobsr\\OneDrive\\Documentos\\GitHub\\ProjetoIntegradorIII\\DATASUS\\Obitos_Procedimento_Municipio\\Tratados\\Obitos_2019.csv', index=False, sep=';', encoding='latin-1')
```
