```python
import pandas as pd
```

```python
# Importando os dados
df_2020_1 = pd.read_csv('C:\\Users\\jobsr\\OneDrive\\Documentos\\GitHub\\IESB\\S3\\PI_3\\Tabelas SUS\\Internacoes por Subgrupo por Municipio\\2020\\2020_1.csv', sep=';', encoding='latin1')
df_2020_2 = pd.read_csv('C:\\Users\\jobsr\\OneDrive\\Documentos\\GitHub\\IESB\\S3\\PI_3\\Tabelas SUS\\Internacoes por Subgrupo por Municipio\\2020\\2020_2.csv', sep=';', encoding='latin1')
df_2020_3 = pd.read_csv('C:\\Users\\jobsr\\OneDrive\\Documentos\\GitHub\\IESB\\S3\\PI_3\\Tabelas SUS\\Internacoes por Subgrupo por Municipio\\2020\\2020_3.csv', sep=';', encoding='latin1')
df_2020_4 = pd.read_csv('C:\\Users\\jobsr\\OneDrive\\Documentos\\GitHub\\IESB\\S3\\PI_3\\Tabelas SUS\\Internacoes por Subgrupo por Municipio\\2020\\2020_4.csv', sep=';', encoding='latin1')
```

```python
# Drop da coluna "Total"
df_2020_1.drop(columns=['Total'], inplace=True)
df_2020_2.drop(columns=['Total'], inplace=True)
df_2020_3.drop(columns=['Total'], inplace=True)
df_2020_4.drop(columns=['Total'], inplace=True)
```

```python
# Gerar uma nova coluna com o ano e o trimestre para cada tabela de 2020
df_2020_1['Ano'] = 2020
df_2020_1['Trimestre'] = 1

df_2020_2['Ano'] = 2020
df_2020_2['Trimestre'] = 2

df_2020_3['Ano'] = 2020
df_2020_3['Trimestre'] = 3

df_2020_4['Ano'] = 2020
df_2020_4['Trimestre'] = 4


# Gerar um código primário unindo as colunas Ano, Trimestre, Código do Município. Essa coluna será usada como chave primária para mesclar as tabelas. Criar uma nova coluna com o nome "UID"
df_2020_1['UID'] = df_2020_1['Ano'].astype(str) + df_2020_1['Trimestre'].astype(str) + df_2020_1['Codigo_Municipio'].astype(str)
df_2020_2['UID'] = df_2020_2['Ano'].astype(str) + df_2020_2['Trimestre'].astype(str) + df_2020_2['Codigo_Municipio'].astype(str)
df_2020_3['UID'] = df_2020_3['Ano'].astype(str) + df_2020_3['Trimestre'].astype(str) + df_2020_3['Codigo_Municipio'].astype(str)
df_2020_4['UID'] = df_2020_4['Ano'].astype(str) + df_2020_4['Trimestre'].astype(str) + df_2020_4['Codigo_Municipio'].astype(str)


# Mesclar as tabelas em uma única tabela
df_2020 = pd.concat([df_2020_1, df_2020_2, df_2020_3, df_2020_4], ignore_index=True)
```

```python
# Mover as três últimas colunas para o início da tabela concatenada do ano
cols = df_2020.columns.tolist()
cols = cols[-3:] + cols[:-3]
df_2020 = df_2020[cols]
```

```python
# Mover as três últimas colunas para o início da tabela do primeiro trimestre
cols = df_2020_1.columns.tolist()
cols = cols[-3:] + cols[:-3]
df_2020_1 = df_2020_1[cols]

# Mover as três últimas colunas para o início da tabela do segundo trimestre
cols = df_2020_2.columns.tolist()
cols = cols[-3:] + cols[:-3]
df_2020_2 = df_2020_2[cols]

# Mover as três últimas colunas para o início da tabela do terceiro trimestre
cols = df_2020_3.columns.tolist()
cols = cols[-3:] + cols[:-3]
df_2020_3 = df_2020_3[cols]

# Mover as três últimas colunas para o início da tabela do quarto trimestre
cols = df_2020_4.columns.tolist()
cols = cols[-3:] + cols[:-3]
df_2020_4 = df_2020_4[cols]
```
